```python



usuarios = {}

def cadastrar():
    """
    Função para cadastrar um novo usuário.
    """
    novo_usuario = input("Digite o novo nome de usuário: ")
    nova_senha = input("Digite a nova senha: ")


    if novo_usuario in usuarios:
        print("Usuário já cadastrado. Tente outro nome de usuário.")
    else:
        usuarios[novo_usuario] = nova_senha
        print("Usuário cadastrado com sucesso.")

def login():
    """
    Função para efetuar login.
    """
    username = input("Digite o nome de usuário: ")
    password = input("Digite a senha: ")


    if username in usuarios and usuarios[username] == password:
        print(f"Bem-vindo, {username}!")
        return username
    else:
        print("Nome de usuário ou senha incorretos.")
        return None

def dev():
    Creditos_Desenvolvedor = ["Juan Pedro Souza de Oliveira", "Felipe Guarniere Pinete", "Matheus Sousa dos Santos"]
    Solicitantes_codigo = ["André Marachini", "Filipe Matos","Guilherme Henrique", "Gustavo Dias","Rhainer Herique"]

    print("Desenvolvedores:")
    for desenvolvedor in Creditos_Desenvolvedor:
        print("-", desenvolvedor)

    print("\nClientes:")
    for cliente in Solicitantes_codigo:
        print("-", cliente)


def menu_login():
    """
    Menu de login e cadastro.
    """

    while True:
        print("╔════════════════════════════════════════╗")
        print("║            Menu de Acesso              ║")
        print("║════════════════════════════════════════║")
        print("║ 1. Login                               ║")
        print("║════════════════════════════════════════║")
        print("║ 2. Cadastro                            ║")
        print("║════════════════════════════════════════║")
        print("║ 3. Créditos                            ║")
        print("╚════════════════════════════════════════╝")

        escolha = input("Escolha uma opção: ")

        if escolha == '1':
            usuario_logado = login()
            if usuario_logado:
                return usuario_logado
        elif escolha == '2':
            cadastrar()
        elif escolha == '3':
              dev()
        else:
            print("Opção inválida, digite novamente.")



def main():
    """
    Função principal.
    """
    usuario_logado = None
    while not usuario_logado:
        usuario_logado = menu_login()

    estoque = {}
    while True:
        print("╔══════════════════════════════════════════════════════════════════════╗")
        print("║                             Bem-vindo                                ║")
        print("║                           Menu de Opções                             ║")
        print("║══════════════════════════════════════════════════════════════════════║")
        print("║ 1. Adicionar Produto                                                 ║")
        print("║══════════════════════════════════════════════════════════════════════║")
        print("║ 2. Verificar Estoque                                                 ║")
        print("║══════════════════════════════════════════════════════════════════════║")
        print("║ 3. Remover Produto                                                   ║")
        print("║══════════════════════════════════════════════════════════════════════║")
        print("║ 4. Editar Quantidade                                                 ║")
        print("║══════════════════════════════════════════════════════════════════════║")
        print("║ 5. Sair                                                              ║")
        print("╚══════════════════════════════════════════════════════════════════════╝")


        escolha = input("Escolha uma opção: ")

        if escolha == '1':
            adicionar_produto(estoque)
        elif escolha == '2':
            verificar_estoque(estoque)
        elif escolha == '3':
            remover_produto(estoque)
        elif escolha == '4':
            edit_quant(estoque)
        elif escolha == '5':
            print("Saindo...")
            break
        else:
            print("Opção inválida, digite novamente.")
def adicionar_produto(estoque):
    """
    Adiciona um produto ao estoque.
    """
    print("╔════════════════════════════════════════════════╗")
    print("║              Adicionar Produto                 ║")
    print("╚════════════════════════════════════════════════╝")
    nome_pessoa = input(" Digite o nome da pessoa (ou 'sair' para voltar): ")
    if nome_pessoa.lower() == 'sair':
        return
    nome_produto = input(" Digite o nome do produto: ")
    quantidade = int(input(" Digite a quantidade: "))

    estoque.setdefault(nome_pessoa, {})
    estoque[nome_pessoa][nome_produto] = estoque[nome_pessoa].get(nome_produto, 0) + quantidade
    print("╔════════════════════════════════════════════════╗")
    print("║   Produto adicionado ao estoque com sucesso.   ║")
    print("╚════════════════════════════════════════════════╝")



def verificar_estoque(estoque):
    """
    Mostra o estoque atual.
    """
    print("╔════════════════════════════════════════════════╗")
    print("║                Estoque Atual                   ║")
    print("╠════════════════════════════════════════════════╣")
    for pessoa, produtos in estoque.items():
        print(f"║ {pessoa}:")
        for produto, quantidade in produtos.items():
            print(f"║   - {produto}: {quantidade}")
    print("╚════════════════════════════════════════════════╝")

def edit_quant(estoque):
    """2
    Remove um produto do estoque.
    """
    print("╔════════════════════════════════════════════════╗")
    print("║                Editar Quantidade               ║")
    print("╚════════════════════════════════════════════════╝")
    nome_pessoa = input(" Digite o nome da pessoa: ")
    nome_produto = input(" Digite o nome do produto que deseja editar a quantidade: ")
    quantidade = int(input(" Digite a quantidade desejada: "))

    if nome_pessoa in estoque and nome_produto in estoque[nome_pessoa]:
        if estoque[nome_pessoa][nome_produto] >= quantidade:
            estoque[nome_pessoa][nome_produto] -= quantidade
            print(f" {quantidade} unidades do produto '{nome_produto}'removidas do estoque de {nome_pessoa}' com sucesso.")
    else:
        print(f" Produto '{nome_produto}' não encontrado no estoque de '{nome_pessoa}'.")

def remover_produto(estoque):
    print("╔════════════════════════════════════════════════╗")
    print("║                Remover Produto                 ║")
    print("╚════════════════════════════════════════════════╝")

    nome_pessoa = input(" Digite o nome da pessoa: ")
    nome_produto = input(" Digite o nome do produto que deseja remover: ")

    if nome_pessoa in estoque and nome_produto in estoque[nome_pessoa]:
        del estoque[nome_pessoa][nome_produto]
        print(f" Produto '{nome_produto}' removido do estoque de '{nome_pessoa}' com sucesso.")
    else:
        print(f" Produto '{nome_produto}' não encontrado no estoque de '{nome_pessoa}'.")
main()
```
